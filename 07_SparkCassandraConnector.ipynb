```python
import pyspark
from pyspark import SparkContext, SparkConf
from pyspark.sql import SparkSession

# create a context including the cassandra connector
conf = pyspark.SparkConf().set('spark.jars.packages', 'datastax:spark-cassandra-connector:2.4.0-s_2.11').set("spark.executor.memory", "512m").setAppName('SparkCass').setMaster('local[1]')
sc = pyspark.SparkContext(conf=conf)
spark = SparkSession(sc)
```

```python
sc
```

Read from a cassandra table

```python
tweetsDf = spark.read.format('org.apache.spark.sql.cassandra').option('spark.cassandra.connection.host', '3.94.217.46')\
                                    .options(table='tweets', keyspace='ksfc_twittercs').load()
```

```python
tweetsDf.count()
```

```python
tweetsDf.take(5)
```

```python
newEggTweets = tweetsDf.filter(tweetsDf["author_id"] == "NeweggService")
print(newEggTweets.count())
newEggTweets.take(5)
```

```python
apologiesDf = tweetsDf.filter(tweetsDf["text"].contains("Sorry"))
apologiesDf.count()
```

```python
apologiesDf.toPandas()
```

```python
apologiesDf.write\
    .format("org.apache.spark.sql.cassandra")\
    .option('spark.cassandra.connection.host', '3.94.217.46')\
    .mode('append')\
    .options(table="sorry_tweets", keyspace="ksfc_twittercs")\
    .save()

#.mode('overwrite').option("confirm.truncate","true")\
```
